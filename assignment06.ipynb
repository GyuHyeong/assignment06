# Import packages for plotting graphs and manipulating data

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Plot the noisy data (x,y1) and clean data (x,y2)

```python

num     = 201
std     = 20
a       = 2 
b       = 10

n       = np.random.rand(num)
nn      = n - np.mean(n)
x       = np.linspace(-100,100,num)
y1      = a * x + nn * std + b 
y2      = a * x + b

plt.plot(x, y1, 'b.', x, y2, 'k--')
plt.show()

# x  : x-coordinate data
# y1 : (noisy) y-coordinate data
# y2 : (clean) y-coordinate data 
# y = f(x) = a * x + b
```

# $\hat{\Theta}$ is solution of least square problem if $\|A\hat{\Theta}-y1\|^2\leq\|A\Theta-y1\|^2$ for any 2-vector $\Theta$.
# We can find $\hat{\Theta}$ by this equation $\hat{\Theta}=(A^TA)^{-1}A^Ty1$

```python
A=np.zeros((num,2))
for i in range(num):
    A[i,0]=1
    A[i,1]=x[i]

theta=((np.linalg.inv(A.T.dot(A))).dot(A.T)).dot(y1)

print(theta)
```

# Plot the line $f(x)=\hat{\Theta}_1+\hat{\Theta}_2x$  with color red.

```python
plt.plot(x, y1, 'b.', x, theta[0]+theta[1]*x, 'r')
plt.show()
```
